```python
# To see interactive  3d display of vacancies and interstitials 
# Instructions: Set picfile to the location of the case you want to see, then run (or play) this cell
# Interaction includes zoom, rotate, stereoscopic view option, and fullscreen option
# The picfile is generated by Fe_all based on the rotation option (0-none, 1-PKA, 2- ellipsoid)
# The default rotation option was PKA, which mean that the PKA (big black sphere) should be at lower Z, somewhat centered on X,Y
# The default colors are black for the interstitials and white for the vacanices
# A larger black dot is used to indicate the location of the PKA hit
# Units are in cell lengths, all xyz scales are the same (e.g., 50x50x50)

import pickle
picfile="/users/lepoire/downloads/005-fpos-10-10.pickle"
with open(picfile, 'rb') as f:
    data = pickle.load(f)
x=data[0,]
y=data[1,]
z=data[2,]
c=data[3,]

import matplotlib 
import ipyvolume as ipv 
import numpy as np
import matplotlib.pyplot as plt
colors = [(1.,1.,1.) if x > 0.5 else (1., 0., 0.) for x in c]
sizes=  [3. if x == 0.5 else 1. for x in c]
ax=plt.gca()
ax.figure.set_size_inches(7,7)
plt.axis('square')
ipv.quickscatter(x, y, z, size=sizes, color=colors, marker="sphere")

```
